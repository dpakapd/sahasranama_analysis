```python
#Import required libraries

from nltk.classify import accuracy #For accuracy predictor
import random #For randomly shuffling the names
import nltk #Natural Language toolkit

file_devi = open("R:/Data for analysis/Sahasranama/Devi.txt","r")
file_vishnu = open("R:/Data for analysis/Sahasranama/Vishnu.txt","r")
devi_content = file_devi.readlines() #Here you are assigning the names in the text file as a 'list'.
vishnu_content = file_vishnu.readlines()

#Adding the label 'Devi' and 'Vishnu' to the corresponding slokas and combining them as 1 list
verse_list = ([(name,'Devi') for name in devi_content]+[(name,'Vishnu') for name in vishnu_content])
print(verse_list)

random.shuffle(verse_list) #Shuffling the names randomly
print("There are %s samples in the dataset." % (len(verse_list)))

#This function is used to extract features of the sloka assigned to the God, especially in the last 3 characters

def name_features(word):
    return {'Feature': word[-3:]} #Extracts the last 3 characters from any string

extract_name_feature = [(name_features(n), name) for (n, name) in verse_list]
print(extract_name_feature)


#Split data into training and test set and use Naive Bayes Classifier

train_set, test_set = extract_name_feature[1000:], extract_name_feature[:1004]

classifier = nltk.NaiveBayesClassifier.train(train_set) #Training the model using Naive Bayes Classifier

#Print the accuracy of the training set
print("The accuracy of the classifier is:", 100*(nltk.classify.accuracy(classifier, train_set)),"%")

#Print the most informative features our model used
classifier.show_most_informative_features(8)
#https://medium.com/@ishan.cdixit/pythons-natural-language-tool-kit-nltk-tutorial-part-3-84a8bcb089bd

#Checking incorrect classification of the model

name_list = verse_list[:2004] #Extract subset to test model output

model_testing = [] #Create an empty list to populate the list of slokas
for (name,god) in name_list:
    model_output = classifier.classify(name_features(name)) #Gives the corresponding God assigned to the output of the function name_features    
    if model_output!=god:
        model_testing.append((god,model_output,name))

for (model_input, model_output, sloka) in model_testing:
    print('Sloka:',sloka,"\n",'Corresponding God:',model_input,"\n",'Model Output:',model_output,"\n")    
```
